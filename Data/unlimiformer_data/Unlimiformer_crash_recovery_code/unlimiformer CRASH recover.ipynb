```python
import re
# The path to your original file
original_file_path = '4clustersoutput.txt'

# The path to the new file where you want to write the extracted text
new_file_path = '4clustersoutputCLEAN.txt'

# Open the original file in read mode
with open(original_file_path, 'r') as original_file:
    # Read the entire content of the file
    file_content = original_file.read()
    
    # Regular expression to find all occurrences of text between '|||' and '</s>', across lines
    pattern = re.compile(r'\|\|\|(.+?)<\/s>', re.DOTALL)
    matches = pattern.findall(file_content)
    
    # Open the new file in write mode
    with open(new_file_path, 'w') as new_file:
        for match in matches:
            # Clean the extracted text to remove leading/trailing whitespaces
            extracted_text = match.strip()
            # Replace newline characters with a space to ensure it's on one line
            single_line_text = extracted_text.replace('\n', ' ')
            # Write the single line text to the new file, followed by a newline character
            new_file.write(single_line_text + '\n')
```

```python
import os

# Directory containing the documents
dir_path = "docs/"

# Iterate over all files in the directory
for filename in os.listdir(dir_path):
    # Check if the file is one of the documents to be removed
    if filename.startswith('q') and filename.endswith('.txt'):
        # Extract the document number from the filename
        doc_number = int(filename[1:-4])  # Removes the 'q' prefix and '.txt' suffix and converts to int
        # Check if the document number is within the range to be removed
        if 0 <= doc_number <= 59:
            # Construct the full path to the file
            file_path = os.path.join(dir_path, filename)
            # Remove the file
            os.remove(file_path)
            print(f"Removed: {filename}")
```
