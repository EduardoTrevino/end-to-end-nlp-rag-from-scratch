```python
import matplotlib.pyplot as plt
import numpy as np

# Range of effect sizes
effect_sizes = np.array([0.2, 0.5, 0.8])
sample_sizes = np.array([analysis.solve_power(effect, power=power, alpha=alpha, ratio=1.0) for effect in effect_sizes])

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(effect_sizes, sample_sizes, marker='o')
plt.title('Sample Size by Effect Size')
plt.xlabel('Effect Size')
plt.ylabel('Sample Size')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from statsmodels.stats.power import TTestIndPower

# parameters
effect_size = 0.5  # medium effect size
alpha = 0.05  # significance level
power = 0.8  # desired power

# create an object
analysis = TTestIndPower()

# calculate sample size
sample_size = analysis.solve_power(effect_size=effect_size, power=power, alpha=alpha, ratio=1.0, alternative='two-sided')
print(f"Required sample size: {np.ceil(sample_size)}")
```
