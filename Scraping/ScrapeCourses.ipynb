```python
import requests
from bs4 import BeautifulSoup

url = "https://enr-apps.as.cmu.edu/assets/SOC/sched_layout_summer_2.htm"
 
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table')
print(table)

f = open("raw.txt", "w")
f.write(response.text)
f.close()
```

```python
rows = table.findAll('tr')
all_cols = []
for i in rows:
    cols = i.find_all("td")
    cols = [j.text for j in cols]
    all_cols.append(cols)
```

```python
cleaned_cols = []

for i in all_cols[2:]:
    
    if i[0] == "\xa0" and len(i) == 10:
        for count in range(3, len(cleaned_cols[-1])):
            cleaned_cols[-1][count].add(i[count])
    elif i[0] != "\xa0" and i[1] != "\xa0":
        newCol = [set() for _ in range(10)]
        for count, j in enumerate(i):
            newCol[count].add(j)

        cleaned_cols.append(newCol)
```

```python
f = open("FinalCourses/summer24_2.txt", "w")

totalCharacters = 0
for i in cleaned_cols:
    text = ""
    text += "Summer 2024\n"
    text += "Course number: {}\n".format(','.join(map(str, i[0])) )
    text += "Title: {}\n".format(','.join(map(str, i[1])) )
    text += "Units: {}\n".format(','.join(map(str, i[2])) )
    text += "Section: {}\n".format(','.join(map(str, i[3])) )
    text += "Days: {}\n".format(','.join(map(str, i[4])) )
    text += "Start: {}\n".format(','.join(map(str, i[5])) )
    text += "End: {}\n".format(','.join(map(str, i[6])) )
    text += "Room: {}\n".format(','.join(map(str, i[7])) )
    text += "Locations: {}\n".format(','.join(map(str, i[8])) )
    text += "Instructors: {}\n".format(','.join(map(str, i[9])))

    totalCharacters += len(text)
    f.write(text + "\n")



f.close()
```
